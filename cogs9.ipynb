```python
import pandas as pd
import matplotlib.pyplot as plt
def conv(s):
    return 1 if s == 'Positive' else 0 if s == 'Neutral' else -1
```

```python
df = pd.read_csv('./Sentiment.csv')
df = df[df.get('candidate') != 'No candidate mentioned'].assign(sentiment=df.get('sentiment').apply(conv))
temp = df.get(['candidate', 'sentiment'])
temp = temp.groupby('candidate').mean()
temp.plot.bar(y='sentiment')
```
